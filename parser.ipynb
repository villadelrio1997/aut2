```python
import os
from pypdf import PdfReader
from dotenv import load_dotenv
import json
load_dotenv()
```

```python
class DecretoParser:
    def __init__(self, pdf_path):
        self.pdf_path = pdf_path
        self.reader = PdfReader(pdf_path)
        self.text = self.extract_text()
    
    def extract_text(self):
        text = ""
        for page in self.reader.pages:
            text += page.extract_text() + "\n"
        return text
    
    def parse(self):
        # Implementar lógica de parsing específica para decretos
        # Esto es un ejemplo muy básico y debe ser adaptado a la estructura real de los decretos
        lines = self.text.splitlines()
        data = {
            "titulo": lines[0] if lines else "",
            "contenido": "\n".join(lines[1:]) if len(lines) > 1 else ""
        }
        return data
    
```

```python
parser = DecretoParser(os.getenv("DECRETO_PATH"))
decreto_data = parser.parse()
```

```python
for key,val in decreto_data.items():
    print(f"{key}: {val[:100]}...")  # Imprime solo los primeros 100 caracteres del contenido para evitar saturar la salida
    print(len(val))
```
